```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
```

```python
web = requests.get("https://www2.2019seagames.com/countries/")
data = BeautifulSoup(web.content, "html.parser")

out = []

for i in data.find_all("em"):
    out.append(i.text)

out
```

```python
country = out[::6]

country = country[0:11]

country
```

```python
gold_raw = out[2::6]

gold_raw
```

```python
gold_raw1 = []
gold = []

for i in range(len(gold_raw)):
    temp = gold_raw[i].split(" ")
    gold_raw1.append(temp[1])
    
for j in range(len(gold_raw1)):
    temp = int(gold_raw1[j])
    gold.append(temp)
    
gold = gold[:11]

gold
```

```python
web1 = requests.get("https://rs.2019seagames.com/RS2019/mobiapp/MedalTally")
data1 = BeautifulSoup(web1.content, "html.parser")

out1 = []

for i in data1.find_all("small"):
    out1.append(i.text)

out1
```

```python
out2 = out1[5:]

out2
```

```python
country1 = out2[::5]

country1
```

```python
gold1 = out2[1::5]

gold1
```

```python
all0 = list(zip(country, gold))
all0.sort()

all1 = list(zip(country1, gold1))
all1.sort()
```

```python
all0
```

```python
all1
```

```python
country_all = []

for i in range(len(all1)):
    temp = all1[i]
    country_all.append(temp[0])

country_all
```

```python
gold_2017 = []

for i in range(len(all0)):
    temp = all0[i]
    gold_2017.append(temp[1])

gold_2017
```

```python
gold_2019 = []

for i in range(len(all1)):
    temp = all1[i]
    gold_2019.append(int(temp[1]))

gold_2019
```

```python
plt.figure(figsize=(15,7))
plt.plot(country_all, gold_2017, "r-o")
plt.plot(country_all, gold_2019, "g-o")
plt.plot(country_all[gold_2017.index(max(gold_2017))], max(gold_2017), '*', markersize = 20, markerfacecolor = "yellow")
plt.plot(country_all[gold_2019.index(max(gold_2019))], max(gold_2019), '*', markersize = 20, markerfacecolor = "yellow")
plt.title('GOLD MEDAL TALLY SEA GAMES 2017 & 2019', fontsize = 30, pad = 60)
plt.xlabel('COUNTRY', fontsize = 20)
plt.ylabel('JUMLAH MEDALI', fontsize = 20)
plt.xticks(rotation = 25)
plt.legend(['Gold 2017', 'Gold 2019', 'Champion 2017', 'Champion 2019'], bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',
           ncol=4, mode="expand", borderaxespad=0.)
plt.grid(True)
plt.show
```

```python
plt.figure(figsize = (15,7))

plt.subplot(1, 2, 1)
plt.pie(gold_2017, labels = country_all, autopct='%1.1f%%')
plt.title('%GOLD MEDAL TALLY SEA GAMES 2017', fontsize = 20)

plt.subplot(1, 2, 2)
plt.pie(gold_2019, labels = country_all, autopct='%1.1f%%')
plt.title('%GOLD MEDAL TALLY SEA GAMES 2019', fontsize = 20)

plt.show()
```
